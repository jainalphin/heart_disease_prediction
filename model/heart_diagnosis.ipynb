```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

```python
df = pd.read_csv('heart.csv')
df
```

```python
df['oldpeak'].value_counts()
```

```python
columns_to_scale=['trestbps','chol','thalach','age','oldpeak']
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df[columns_to_scale]= scaler.fit_transform(df[columns_to_scale])
```

```python
df
```

```python
s=scaler.transform([[145,233,150,63,2.3]])
s[0][0]
```

```python
X=df.drop('target',axis=1)
y=df['target']
```

```python
X.shape
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score,confusion_matrix
randomforest_classifier= RandomForestClassifier(n_estimators=10)

score=randomforest_classifier.fit(X_train,y_train)
y_pred= randomforest_classifier.predict(X_test)
randomforest_classifier_score=accuracy_score(y_test,y_pred)
```

### LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
logistic_Regression_model= LogisticRegression()
logistic_Regression_model.fit(X_train,y_train)
y_pred =logistic_Regression_model.predict(X_test)
print(accuracy_score(y_test,y_pred))
logistic_Regression_score=accuracy_score(y_test,y_pred)
```

### DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier

dt_model= DecisionTreeClassifier()
dt_model.fit(X_train,y_train)

y_pred =dt_model.predict(X_test)
print(accuracy_score(y_test,y_pred))
dt_score=accuracy_score(y_test,y_pred)
```

###  GaussianNB

```python
from sklearn.naive_bayes import GaussianNB

gaussian_naive_bayes= GaussianNB()
gaussian_naive_bayes.fit(X_train,y_train)
y_pred=gaussian_naive_bayes.predict(X_test)
print(accuracy_score(y_test,y_pred))
nb_score=accuracy_score(y_test,y_pred)
```

###  SVC

```python
from sklearn.svm import SVC

svc= SVC()
svc.fit(X_train,y_train)
y_pred=svc.predict(X_test)
print(accuracy_score(y_test,y_pred))
svc_score=accuracy_score(y_test,y_pred)
```

### KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn= KNeighborsClassifier(n_neighbors=10)
knn.fit(X_train,y_train)
y_pred=knn.predict(X_test)
print(accuracy_score(y_test,y_pred))
knn_score=accuracy_score(y_test,y_pred)
```

```python
model_comparision=pd.DataFrame({
    'Model': ['Logistic Regression','K-Nearest Neighbor', 'Decision Tree',
                 'Random Forest','Naive Bayes','SVC'],
    'Score':[logistic_Regression_score, knn_score, dt_score, 
              randomforest_classifier_score, nb_score,svc_score]})
```

# HyperParamer tunnig

```python
# Random search Cv

#number of tree in Random Searcv
n_estimators =[int(i) for i in np.linspace(start=100,stop=1200,num=12)]

#Number of fearures to consider at every split
max_fearures=['auto','sqrt']

# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]

# max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]

# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]

# Method of selecting samples for training each tree
# bootstrap = [True, False]
```

```python
random_grid ={
    'n_estimators':n_estimators,
    'max_features':max_fearures,
    'max_depth':max_depth,
    'min_samples_split':min_samples_split,
    'min_samples_leaf':min_samples_leaf
}
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations
rf_random = RandomizedSearchCV(estimator = RandomForestClassifier(), 
                               param_distributions = random_grid,
                               scoring='roc_auc_ovo', 
                               n_iter = 50, 
                               cv = 5, 
                               verbose=2,
                               random_state=42, 
                               n_jobs = 1)
rf_random.fit(X,y)
```

```python
rf_random.best_score_
```
